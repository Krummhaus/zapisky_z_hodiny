```python
import numpy as np
import pandas as pd
```

```python
# Načteme data bez hlavičky a nechame pandas nastavit index automaticky
link  = 'https://www.czso.cz/documents/10180/32853391/300002164403.xlsx/1fed9fb2-4dbf-4b0f-bfbe-4c2d3202a18f?version=1.1'
data = pd.read_excel(link, head = None, skiprows=5)
data.head()
```

```python
data.iloc[2]
```

```python
# podíváme se na první řádek
data.iloc[0]
```

```python
data.iloc[2].fillna(method = 'ffill')
```

```python
# Máme sloučené buňky, musíme proto doplnit hodnoty sloupců
col1 = data.iloc[0].fillna(method = 'ffill')
col2 = data.iloc[2].fillna(method = 'ffill').str.split(pat='/', n = 0).str.get(0)
col3 = data.iloc[3].fillna(method = 'ffill')
columns = [col1, col2, col3]
columns = list(zip(*columns))
columns
```

```python
data.iloc[7]
```

```python
# spojene názvy sloupců dáme jako hlavičku
# řádky, které jsme k tomu potřebovali, odstraníme 
data.columns = columns
data = data.drop([0,1,2,3,4,7])
data.head()
```

```python
# v tabulce se nám střídají prázdné sloupce a sloupce s daty
# odstraníme prázdné sloupce 
# sloupec musí obsahovat minimálně 6 nenulových hodnot
data = data.dropna(axis=1, how='all', thresh = 6)
data.head()
```

```python
# U prvního sloupce vyměníme nan za 'Země'
data = data.rename(columns={('Země', np.nan, np.nan) : ('Země', 'Země', 'Země'), ('Country', '65 let a více ', 'muži') : ('Country', 'Country', 'Country')})
data.head()
```

```python
data
```

```python
# Odstraníme všechny koncové řádky, které nepotřebujeme
data = data.drop(data.index[data.index > 40])
data
```

```python
# Odstraníme 1) u Lucemburska
data.replace('Lucembursko1)', 'Lucembursko', inplace = True)
data
```

```python
# Odstraníme řádek 'Ostatní země'
data.drop(36, inplace = True)
data
```

```python
# Změníme '.' na NaN
data.replace('.', np.nan, inplace = True)
data
```
